## Ablauf

1. Spezialisten als Tools registrieren
2. Hauptagent wählt Tool anhand Aufgabe
3. Spezialist liefert begrenztes Ergebnis

```python
import subprocess, sys

prompt = "Use a research specialist and a coding specialist as callable tools."
cmd = ['agent-patterns', 'run', "agents-as-tools", prompt]
print('$ ' + ' '.join(cmd))
completed = subprocess.run(cmd, text=True, capture_output=True)
print(completed.stdout)
if completed.stderr:
    print(completed.stderr, file=sys.stderr)
```
