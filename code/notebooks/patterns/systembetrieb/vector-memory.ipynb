## Modellzugang in Colab

GitHub oder Colab stellen keinen Model-Token automatisch bereit. Ohne eigenen Token läuft die Demo im erklärenden Offline-Modus. Für echte Modellantworten kannst du GitHub Models nutzen, weil der Endpoint OpenAI-kompatibel ist. Füge bei Bedarf vor dem Demo-Run eine Code-Zelle mit diesem Inhalt ein:

```python
import getpass
import os

os.environ["AGENT_PROVIDER"] = "github"
os.environ["GITHUB_TOKEN"] = getpass.getpass("GitHub Token mit Models-Zugriff: ")
os.environ["GITHUB_MODEL"] = "openai/gpt-4.1-mini"
```

Der Token braucht Zugriff auf GitHub Models, in der Regel `models:read`.
